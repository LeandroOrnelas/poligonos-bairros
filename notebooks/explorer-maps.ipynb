```python
import folium
from pathlib import Path
import geopandas as gpd
from branca.colormap import linear

```

```python
geojson_path = Path("data") / "bairros_uberlandia.geojson"

# Caminho do arquivo .geojson
geojson_path = Path("..") / "data" / "bairros_uberlandia.geojson"
gdf_bairros = gpd.read_file(geojson_path)
```

```python

# Criar o mapa centrado
mapa = folium.Map(location=[-18.914, -48.275], zoom_start=12)

# Adicionar os bairros coloridos conforme a coluna 'cor'
folium.GeoJson(
    gdf_bairros,
    name="Bairros coloridos por zona",
    style_function=lambda feature: {
        "fillColor": feature["properties"]["cor"],
        "color": "black",
        "weight": 0.5,
        "fillOpacity": 0.6,
    },
    tooltip=folium.GeoJsonTooltip(fields=["bairro", "zona_territorial"])
).add_to(mapa)

# Exibir
mapa
```
